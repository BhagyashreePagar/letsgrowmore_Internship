## import the libraries 

```python
import cv2
import matplotlib.pyplot as plt #use to show images
```

# Read in the image

```python
img=cv2.imread("cats.jpg")
```

# show the image

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
#Convert the image to gray scale
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
```

# invert the image

```python
inv =255-gray
plt.imshow(inv,cmap='gray')
```

# blur image using gaussianBlur function

```python
blur=cv2.GaussianBlur(inv,ksize=(21,21),sigmaX=0,sigmaY=0)
plt.imshow(blur,cmap='gray')
```

# invert blurred image

```python
inv_blur=255-blur
plt.imshow(inv_blur,cmap='gray')
```

# create pencil sketch img

```python
pencil_sketch_img=cv2.divide(gray,inv_blur,scale=256.0)
plt.imshow(pencil_sketch_img,cmap='gray')
```

```python
#method 2 for converting image
dodge= lambda image,mask:cv2.divide(image,255-mask,scale=256)
```

```python
blended=dodge(gray,blur)
```

```python
plt.imshow(blended,cmap='gray')
```
